```python
# Загружаем данные за период 01.01.2018 - 23.12.2019
usdrub_df = pd.read_csv('/Users/ekaterina/Desktop/w/нис/USD000UTSTOM_180101_191223.txt',
                       sep=';',
                       usecols = ['<DATE>', '<CLOSE>'],
                       parse_dates=['<DATE>'], dayfirst=True)

eurrub_df = pd.read_csv('/Users/ekaterina/Desktop/w/нис/EUR_RUB__TOM_180101_191223.txt',
                       sep=';',
                       usecols = ['<DATE>', '<CLOSE>'],
                       parse_dates=['<DATE>'], dayfirst=True)

interbankru_df = pd.read_csv('/Users/ekaterina/Desktop/w/нис/interrate_ru.csv',
                            sep=';',
                            parse_dates=['<DATE>'], dayfirst=True)

interbankus_df = pd.read_csv('/Users/ekaterina/Desktop/w/нис/interrate_us.csv',
                            sep=';',
                            parse_dates=['<DATE>'], dayfirst=True)

interbankeu_df = pd.read_csv('/Users/ekaterina/Desktop/w/нис/interrate_eu.csv',
                            sep=';',
                            parse_dates=['<DATE>'], dayfirst=True)
```

```python
# Объединяем данные в один датафрейм
df = pd.merge(usdrub_df, eurrub_df,
                  how = 'inner',
                  left_on = '<DATE>',
                  right_on = '<DATE>')

df = pd.merge(df, interbankru_df,
                  how = 'inner',
                  left_on = '<DATE>',
                  right_on = '<DATE>')

df = pd.merge(df, interbankus_df,
                  how = 'inner',
                  left_on = '<DATE>',
                  right_on = '<DATE>')

df = pd.merge(df, interbankeu_df,
                  how = 'inner',
                  left_on = '<DATE>',
                  right_on = '<DATE>')

# Меняем названия столбцов
df = df.rename (columns = {'<DATE>':'date', '<CLOSE>_x':'usdrub', '<CLOSE>_y':'eurrub', '<RATE>_x':'rate_ru', '<RATE>_y':'rate_us', '<RATE>':'rate_eu'})

# Добавляем столбец с указанием количества дней до экспирации
df['counter'] = range(len(df))
df['till_exp'] = 493 - df['counter']
del df['counter']
```

Оценка стоимости форвардов на каждый момент времени производится согласно формуле оценки валютных форвардов:

$$F_0 = S_0*e^{(r_i-r_o)*T}$$

где $S_0$ - цена базового актива;    
    $r_i$ - внутренняя ставка;  
    $r_o$ - внешняя ставка;  
    $T$ - отношение количества дней до экспирации к общему количеству дней жизни контракта.

```python
# Рассчитываем текущую цену форварда
df['fo_usdrub'] = df['usdrub']*np.exp((df['rate_ru'] - df['rate_us'])*df['till_exp']/494/100)
df['fo_eurrub'] = df['eurrub']*np.exp((df['rate_ru'] - df['rate_eu'])*df['till_exp']/494/100)
```

```python
# Рисунок для котировок форварда на usdrub
fig, ax = plt.subplots()
plt.grid()
plt.title('Форвард рубль-доллар')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.axvline(dt.datetime(2019, 12, 31), color = "red")
ax.plot(df.date, df.fo_usdrub, color = "blue", label = "fo_usdrub")
plt.xticks(rotation = 75) 
ax.legend()
plt.show()
```

```python
# Рисунок для котировок форварда на eurrub
fig, ax = plt.subplots()
plt.grid()
plt.title('Форвард рубль-евро')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.axvline(dt.datetime(2019, 12, 31), color = "red")
ax.plot(df.date, df.fo_eurrub, color = "green", label = "fo_eurrub")
plt.xticks(rotation = 75) 
ax.legend()
plt.show()
```

Оценка математического ожидания и стандартного отклонения портфеля производится согласно формуле:  
$$mean_{portfolio} = \sum_{i=1}^n w_i*mean_i$$  
$$std_{portfolio} = \sqrt{\sum_{i=1}^n {w_i^2*std_i^2} + \sum_{i=1}^n \sum_{i<>j}^n {cov(r_i,r_j)}}$$  
где $w_i$ - вес инструмента i в портфеле  
$mean_i$ - математическое ожидание инструмента i в портфеле  
$std_i$ - стандартное отклонение инструмента i в портфеле  
$cov(r_i,r_j)$ - ковариация цен инструментов в портфеле

```python
# Расчетываем исторические показатели портфеля из двух форвардов
mean_usd = df.mean()['fo_usdrub']
mean_eur = df.mean()['fo_eurrub']
std_usd = df.std()['fo_usdrub']
std_eur = df.std()['fo_eurrub']
cov = np.cov(df.fo_usdrub,df.fo_eurrub)[0][1]
w1 = 0.5
w2 = 0.5
mean_portfolio = w1*mean_usd + w2*mean_eur
std_portfolio = w1*w1*std_usd*std_usd + w2*w2*std_eur*std_eur + 2*w1*w2*cov
```

Оценка стоимости опциона call на каждый момент времени производится согласно формуле Мертона-Блека-Шоулза:  
$$ C(S,t) = SN(d_1) - Ke^{-r(T-t)}N(d_2)$$  
где,  
$$d_1 = \frac{ln(\frac{S}{K})+\frac{r+{\sigma^2/2}}{T-t}}{\sigma\sqrt{T-t}}$$  
$$d_2 = d_1 - \sigma\sqrt{T-t}$$  
$C(S,t)$ - текущая стоимость опциона call в момент t до истечения срока опциона;  
$S$ - текущая цена базисного актива;  
$N(x)$ - вероятность того, что отклонение будет меньше в условиях стандартного нормального распределения;  
$K$ - цена исполнения опциона;  
$r$ - безрисковая процентная ставка;  
$T-t$ - время до истечения срока опциона (период опциона);  
$\sigma$ - волатильность базового актива.
